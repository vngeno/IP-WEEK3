```python
#loading Telcom_dataset3
df = pd.read_csv('Telcom_dataset3.csv')
df
```

```python
df.duplicated()
```

```python
#checking if there's missing data
df.isnull()
```

```python
#dropping missing data 
df_no_missing = df.dropna()
df_no_missing
```

```python
#dropping duplicates
df.drop_duplicates()
#dropping 'NA' cells in rows
df_cleaned = df.dropna(how='all')
df_cleaned
#dropping column if it contains missing values
df.dropna(axis=1, how='all')
# Filling in missing data with zero
df.fillna(0)
df.head()
```

```python
#Create a groupby variable that groups value by product. This is using the telecom dataset2
groupby_product = df['VALUE'].groupby(df['PRODUCT'])
groupby_product
```

```python
#Descriptive statistics by group
df['VALUE'].groupby(df['PRODUCT']).describe()
```

```python
# Mean of each PRODUCT’s VALUE
groupby_product.mean()
```

```python
# Mean preTestScores grouped by PRODUCT and CELL_ON_SITE
df['VALUE'].groupby([df['PRODUCT'], df['CELL_ON_SITE']]).mean()
```

```python
# Create a pivot table of group means, by CELL_ON_SITE and PRODUCT
pd.pivot_table(df, index=['PRODUCT','CELL_ON_SITE'], aggfunc='mean')
```
